```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
data_xls = pd.read_excel('NewHealthyData.xlsx', 'Sheet1', usecols="B:I", nrows=344)
data_xls.to_csv('healthyData.csv', encoding='utf-8', index=False)
```

```python
data_xls = pd.read_excel('Schizont data.xlsx', 'schizont', usecols="B:I", nrows=344)
data_xls.to_csv('schizontData.csv', encoding='utf-8', index=False)
```

```python
healthy_data = pd.read_csv('healthyData.csv')
schizont_data = pd.read_csv('schizontData.csv')

healthy_perimeter = healthy_data['perimeter']
schizont_perimeter = schizont_data['Perimeter to area ratio']

plt.figure(figsize=(20,10))
plt.hist(healthy_perimeter, bins=100, range= [0.75,2], alpha=0.7, color='lime', density=True)
plt.hist(schizont_perimeter, bins=100, range=[0.75, 2], alpha=0.7, color='crimson', density=True)

sch_per_mu = np.mean(schizont_perimeter)
sch_per_std = np.std(schizont_perimeter)
hea_per_mu = np.mean(healthy_perimeter)
hea_per_std = np.std(healthy_perimeter)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
s_per = norm.pdf(x, sch_per_mu, sch_per_std)
h_per = norm.pdf(x, hea_per_mu, hea_per_std)


plt.plot(x, s_per, 'k', linewidth=2, color='blue')
plt.plot(x, h_per, 'k', linewidth=2)


plt.legend(['healthy distribution', 'schizont distribution', 'healthy', 'schizont'])
plt.show()
```

```python
array = np.asanyarray(healthy_perimeter)
sort_ar = np.sort(array)
sort_ar = sort_ar[:-1]

array2 = np.asanyarray(schizont_perimeter)
plt.figure(figsize=(20,10))
ax1 = plt.subplot(131)
ax1.boxplot(array)
ax1.set_title('Healthy Data')

ax = plt.subplot(132)
ax.boxplot(sort_ar)
ax.set_title('Healthy Data without outlier')

ax2 = plt.subplot(133)
ax2.boxplot(array2)
ax2.set_title('Schizont Data')
```
