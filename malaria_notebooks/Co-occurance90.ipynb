```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
data_xls = pd.read_excel('Healthy(90).xlsx', 'Sheet1', nrows=344)
data_xls.to_csv('healthyData(90).csv', encoding='utf-8', index=False)

data_xls = pd.read_excel('Schizont(90).xlsx', 'Sheet1', nrows=344)
data_xls.to_csv('schizontData(90).csv', encoding='utf-8', index=False)
```

```python
healthy_data = pd.read_csv('healthyData(90).csv')
schizont_data = pd.read_csv('schizontData(90).csv')
```

```python
healthy_contrast = healthy_data['contrast']
schizont_contrast = schizont_data['contrast']
#Stack the data
plt.figure(figsize=(20,10))
plt.hist(healthy_contrast, bins=100, alpha=0.7, color='lime', density=True)
plt.hist(schizont_contrast, bins=100, alpha=0.7, color='crimson', density=True)

sch_con_mu = np.mean(schizont_contrast)
sch_con_std = np.std(schizont_contrast)
hea_con_mu = np.mean(healthy_contrast)
hea_con_std = np.std(healthy_contrast)

con_xmin, con_xmax = plt.xlim()
x_con = np.linspace(con_xmin, con_xmax, 100)
sch_con = norm.pdf(x_con, sch_con_mu, sch_con_std)
hea_con = norm.pdf(x_con, hea_con_mu, hea_con_std)


plt.plot(x_con, sch_con, 'k', linewidth=2, color='blue')
plt.plot(x_con, hea_con, 'k', linewidth=2)

plt.legend(['healthy distribution', 'schizont distribution', 'healthy', 'schizont'])
plt.show()
```
